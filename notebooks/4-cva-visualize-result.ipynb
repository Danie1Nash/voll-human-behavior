```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

data = [
    [ 72, 20, 4, 4],     # Блок
    [ 40, 43, 10, 7],    # Прием мяча снизу
    [ 70, 15, 10, 5],    # Падение
    [ 66, 28, 3, 3],     # Прыжок
    [ 65, 25, 2, 8],     # Передвижение
    [ 65, 24, 5, 6],     # Прием мяча сверху
    [ 55, 26, 11, 7],     # Атакующий удар
    [ 63, 14, 16, 7],    # Стойка на месте
    [ 50, 39, 1, 9],     # Ожидание
]
```

```python
data = np.array(data)/100
```

```python
labels =   ["Блок",
            "Прием мяча снизу",
            "Падение",
            "Прыжок",
            "Передвижение",
            "Прием мяча сверху",
            "Атакующий удар",
            "Стойка на месте",
            "Ожидание"]
```

```python
plt.figure(figsize=(5, 8), dpi=80)

df_cm = pd.DataFrame(data, index=labels, columns=["1", "2", "3", "4"])
heatmap = sns.heatmap(df_cm, annot_kws={"size": 18}, annot=True, fmt=".2f", cmap="Greens")

heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right',fontsize=15)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right',fontsize=15)
plt.ylabel('Тип действия')
plt.xlabel('Результат работы')
plt.savefig('confusion_matrix')
```

```python
data = [
    [77, 72, 20, 4, 4],     # Блок
    [43, 40, 43, 10, 7],    # Прием мяча снизу
    [68, 70, 15, 10, 5],    # Падение
    [72, 66, 28, 3, 3],     # Прыжок
    [70, 65, 25, 2, 8],     # Передвижение
    [72, 65, 24, 5, 6],     # Прием мяча сверху
    [57, 55, 26, 11, 7],     # Атакующий удар
    [65, 63, 14, 16, 7],    # Стойка на месте
    [57, 50, 39, 1, 9],     # Ожидание
]
```

```python
data = np.array(data)
```

```python
data
```

```python
acc = data[:, 1] / np.sum(data[:, 1:4], axis = 1) * 100
```

```python
res = np.concatenate((data[:, 0].reshape(-1, 1, ), acc.reshape(-1, 1)), axis = 1)
```

```python
res
```

```python
plt.figure(figsize=(4, 8), dpi=80)

df_cm = pd.DataFrame(res, index=labels, columns=["До интегации", "После интеграции"])
heatmap = sns.heatmap(df_cm, annot_kws={"size": 18}, annot=True, fmt=".2f", cmap="Greens")

heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right',fontsize=15)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right',fontsize=15)
figure = heatmap.get_figure()
plt.ylabel('Тип действия')
plt.xlabel('Результат работы')
figure.savefig('confusion_matrix')
plt.savefig("test.svg")
```
